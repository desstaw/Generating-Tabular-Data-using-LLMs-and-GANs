```python
import numpy as np
import pandas as pd
import logging
import matplotlib.pyplot as plt
from sklearn import datasets
from be_great import GReaT

url = 'https://raw.githubusercontent.com/desstaw/Generating-Tabular-Data-using-LLMs-and-GANs/main/heart_disease.csv'
data = pd.read_csv(url)
```

```python
# Mapping chest pain type
mapping_chestPain = {
    'TA': 'Typical Angina',
    'ATA': 'Atypical Angina',
    'NAP': 'Non-Anginal Pain',
    'ASY': 'Asymptomatic'
}

data['ChestPainType'] = data['ChestPainType'].replace(mapping_chestPain)

new_column_name = 'ChestPainType'
data = data.rename(columns={'ChestPainType': new_column_name})

# Mapping Sex
mapping_sex = {
    'M': 'male',
    'F': 'female'
}

data['Sex'] = data['Sex'].replace(mapping_sex)

# Mapping fasting blood sugar
mapping_FastingBS = {
    1: 'FastingBS > 120 mg/dl',
    0: 'FastingBS <= 120 mg/dl'
}

data['FastingBS'] = data['FastingBS'].replace(mapping_FastingBS)

new_column_name = 'FastingBloodSugar'
data = data.rename(columns={'FastingBS': new_column_name})

# Mapping Resting ECG
mapping_RestingECG = {
    'ST': 'ST-T wave abnormality',
    'LVH': 'probable or definite left ventricular hypertrophy by Estes criteria'
}

data['RestingECG'] = data['RestingECG'].replace(mapping_RestingECG)

# Mapping MaxHR
new_column_name = 'MaxHeartRate'
data = data.rename(columns={'MaxHR': new_column_name})

# Mapping resting blood pressure
new_column_name = 'RestingBloodPressure'
data = data.rename(columns={'RestingBP': new_column_name})

# Mapping Exercise Angina
mapping_ExerciseAngina = {
    'Y': 'Yes',
    'N': 'No'
}

data['ExerciseAngina'] = data['ExerciseAngina'].replace(mapping_ExerciseAngina)

#Mapping ST slope
new_column_name = 'PeakExerciseSlope'
data = data.rename(columns={'ST_Slope': new_column_name})

# Mapping Heart Disease
mapping_HeartDisease = {
    1: 'Present',
    0: 'Absent'
}

data['HeartDisease'] = data['HeartDisease'].replace(mapping_HeartDisease).astype(str)
```

```python
categorical_features = ['Sex', 'ChestPainType', 'FastingBloodSugar', 'RestingECG', 'ExerciseAngina', 'PeakExerciseSlope', 'HeartDisease']
```

```python
great = GReaT("distilgpt2",
              epochs=100,
              save_steps=2000,
              experiment_dir="trainer_heart",
             )

trainer = great.fit(data)
```
